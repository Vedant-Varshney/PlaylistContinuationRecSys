## Data Characteristics

1. How many unique songs are there?
2. How many unique playlists are there?
3. What is the distribution for the number of songs per playlist?

```python
df.count()
```

```python
uniques = pd.Series({col: len(df[col].unique()) for col in df.columns})
uniques
```

Approx. 10,000 unique playlists

Approx. 105,000 unique songs

There is also a high degree of cardinality to artist name and album. We should explore their distributions to see whether we need to retain all categories or if we can take the top k categories and keep most of the data.

### Songs per playlist distribution

```python
songs_per_playlist = pd.value_counts(df.playlist_id)
```

```python
plt.figure(figsize=(12,10))

plt.xlabel("Playlists")
plt.ylabel("Songs per Playlist")

plt.plot(songs_per_playlist.values, marker="x")
plt.show()
```

There seems to be a very long tail in the number of songs per playlist.

A minority of playlists contain several thousand songs but the vast majority contain comparatively very few.

How many playlists do not have at least 15 songs?

Note - 15 is an arbitr. low number

```python
songs_threshold = 15
very_short_playlists = [x for x in songs_per_playlist.values if x < songs_threshold]
len(very_short_playlists)
```

```python
len(very_short_playlists)/uniques.playlist_id
```

Approx. 45% of all playlists contain less than 15 songs

```python
songs_value_counts = pd.value_counts(df.playlist_id)
```

```python
%matplotlib notebook

plt.xlabel("Songs per Playlist")
plt.ylabel("Number of Playlists")
songs_value_counts.hist(bins=1000)
```

```python
len([x for x in songs_per_playlist if x<5])/uniques.playlist_id
```

Approx. 10% of all playlists contain less than 5 songs

The dataset is not ideal however more than 50% of playlists do have a reasonable number of songs per playlist (>15).

**We may have to remove the very short playlists and truncate the very long playlists later on.**

### Cardinality of artist name and album

```python
df
```

```python
uniques
```

```python
def plot_value_occurences(column):
    value_counts = pd.value_counts(df[column])

    plt.figure(figsize=(12,10))

    plt.xlabel(column)
    plt.ylabel(f"{column} Occurences")

    plt.plot(value_counts.values, marker="x")
    plt.show()
```

```python
%matplotlib inline

plot_value_occurences("artist_name")
```

```python
%matplotlib inline

plot_value_occurences("album")
```

Again, we see a long tail for both the artists and albums. This suggests we likely have to retain the high cardinality and thus will need embedding layers in our model.

```python
df.to_csv("cleaned_data_0.2.csv")
```

## Popularity

## Tags

Each tag in the above dataset consists of a list of genres; i.e. a single tag is not a single genre but rather a list of genres.

We must change the tags column such that all permutations of the same list of genres have the exact same tag.

```python
df = pd.read_csv("cleaned_data_0.2.csv")
```

```python
print(df.tags.unique())
print(f"length: {len(df.tags.unique())}")
```

```python
set_tags = pd.Series([",".join(set(tag.lower().split(","))) for tag in df.tags])
tags = set_tags.unique()
```

```python
print(len(set_tags))
print(len(tags))
```

```python
set_tags
```

```python
df["tags"] = set_tags
```

```python
assert not pd.isna(df).any().any()
```
